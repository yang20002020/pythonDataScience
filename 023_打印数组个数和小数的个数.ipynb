```python
import numpy as np
np.arange(100)
```

```python
import sys
np.set_printoptions(threshold=sys.maxsize) #完全显示 所有元素
np.arange(100)
```

```python
np.linspace(0,10,100)
```

```python
np.set_printoptions(precision=3)
np.linspace(0,10,100)
```

```python
np.set_printoptions(threshold=10) # 不会打印所有的元素
np.linspace(0,10,100)
```
