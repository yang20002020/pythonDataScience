```python
import numpy as np
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'
wid = np.genfromtxt(url, delimiter=',', dtype='float', usecols=[1]) # 获取第一列数据
wid
```

```python
wid.shape
```

```python
smax = np.max(wid)
smin = np.min(wid)
s = (wid - smin) / (smax-smin)
s
```
